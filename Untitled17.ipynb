```python
import io
import pandas as pd
import requests  # Install this if you don't have it: `pip install requests`

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv"

# Fetch the CSV data
response = requests.get(URL)
# Convert to a file-like object for pandas
text = io.BytesIO(response.content)
df_can = pd.read_csv(text)

print('Data read into a pandas dataframe!')
```

```python
#Line Plot
#A line plot displays the relationship between two continuous variables over a continuous interval, showing the trend or pattern of the data.

#Let's created a line plot to visualize the immigrants (to Canada) trend during 1980 to 2013.
#We need the Total of year-wise immigrants,

#We will create a new dataframe for only columns containing the years
#then, we will apply sum() on the dataframe

#You can do create a line plot directly on axes by calling plot function plot()
```

```python
#As years is in the array format, you will be required to map it to str for plotting
#y=list(map(str, years))

#creating df with only years columns from 1980 - 2013
df_line=df_can[years]

#Applying sum to get total immigrants year-wise
total_immigrants=df_line.sum()
total_immigrants
```

```python
#Create figure and axes
fig, ax = plt.subplots()

#plot the line
ax.plot(total_immigrants)

#Setting uo the Title
ax.set_title('Immigrant between 1980 to 2013')

#Setting up the Labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')

#Display the plot
plt.show()
```

```python
# Calculate total immigration for 1980-2013
years = [str(year) for year in range(1980, 2014)]
df_can['Total'] = df_can[years].sum(axis=1)

# Sort and get bottom 5 countries
df_can.sort_values('Total', ascending=True, inplace=True)
bottom_5 = df_can.head(5)

# Create the plot
plt.figure(figsize  =  (10, 6))
plt.bar(bottom_5['Country'], bottom_5['Total'], color  =  'lightcoral')

# Formatting
plt.title('5 Countries with Least Immigration to Canada (1980-2013)')
plt.xlabel('Country')
plt.ylabel('Number of Immigrants')
plt.xticks(rotation  =  45)
plt.grid(axis  =  'y', linestyle  =  '--', alpha  =  0.4)

# Add value labels
for i, value in enumerate(bottom_5['Total']):
    plt.text(i, value, f'{int(value)}', ha  =  'center', va  =  'bottom')

plt.tight_layout()
plt.show()
```

```python
#Histogram
#A histogram is a way of representing the frequency distribution of numeric dataset. The way it works is it partitions the x-axis into bins, assigns each data point in our dataset to a bin, and then counts the number of data points that have been assigned to each bin. So the y-axis is the frequency or the number of data points in each bin. Note that we can change the bin size and usually one needs to tweak it so that the distribution is displayed nicely.

#Let's find out the frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013?
```

```python
df_country = df_can.groupby(['Country'])['2013'].sum().reset_index()

#Create figure and axes
fig, ax = plt.subplots(figsize=(10, 4))
ax.hist(df_country['2013'])
ax.set_title('New Immigrants in 2013')
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Countries')
ax.legend(['Immigrants'])

#Display the plot
plt.show()
```

```python
# Plot the bar
fig, ax = plt.subplots(figsize=(10, 4))
count = ax.hist(df_country['2013'])

#you can check the arrays in count with indexing count[0] for count, count[1] for bins

ax.set_title('New Immigrants in 2013')
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Countries')
ax.set_xticks(list(map(int,count[1])))
ax.legend(['Immigrants'])

#Display the plot
plt.show()
```

```python
# Question: What is the immigration distribution for China and India for years 2000 to 2013?
```

```python
# Filter data for China and India (2000-2013)
years = [str(year) for year in range(2000, 2014)]
countries = ['China', 'India']
df_filtered = df_can[df_can['Country'].isin(countries)][['Country'] + years]

# Melt the dataframe for plotting
df_melted = df_filtered.melt(id_vars='Country', var_name='Year', value_name='Immigrants')

# Plotting
plt.figure(figsize=(12, 6))

# Line plot version
plt.subplot(1, 2, 1)
for country in countries:
    data = df_melted[df_melted['Country'] == country]
    plt.plot(data['Year'], data['Immigrants'], '-o', label=country)
plt.title('Immigration Trend (2000-2013)')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.xticks(rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Bar plot version (FIXED SECTION)
plt.subplot(1, 2, 2)
width = 0.4
x = range(len(years))
for i, country in enumerate(countries):
    data = df_filtered[df_filtered['Country'] == country][years].values.flatten()  # Fixed line
    plt.bar([xi + width*i for xi in x], data, width, label=country)
plt.title('Yearly Comparison')
plt.xlabel('Year')
plt.xticks([xi + width/2 for xi in x], years, rotation=45)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
#Sub-plotting
#Let us explore how to display more than one plot on the same figure and specify the number of rows and columns to be created to the subplots function.

#For instance, let’s create a line and scatter plot in one row
#plt.subplots()
#You can use the same functions using which you plotte lne and scatter plots at the start of this lab
#Both the subplots will be sharing the same y-axis as the data in the y-axis is the same. So, assign the ‘Sharey’ parameter as True in the code below. Also notice the use of 'suptitle'

```

```python
# Create a figure with two axes in a row

fig, axs = plt.subplots(1, 2, sharey=True)

#Plotting in first axes - the left one
axs[0].plot(total_immigrants)
axs[0].set_title("Line plot on immigrants")

#Plotting in second axes - the right one
axs[1].scatter(total_immigrants.index, total_immigrants)
axs[1].set_title("Scatter plot on immigrants")

axs[0].set_ylabel("Number of Immigrants")

#Adding a Title for the Overall Figure
fig.suptitle('Subplotting Example', fontsize=15)

# Adjust spacing between subplots
fig.tight_layout()


# Show the figure
plt.show()
```

```python
#You can also implement the subplotting with add_subplot() as below:-
```

```python
# Create a figure with Four axes - two rows, two columns
fig = plt.figure(figsize=(8,4))

# Add the first subplot (top-left)
axs1 = fig.add_subplot(1, 2, 1)
#Plotting in first axes - the left one
axs1.plot(total_immigrants)
axs1.set_title("Line plot on immigrants")

# Add the second subplot (top-right)
axs2 = fig.add_subplot(1, 2, 2)
#Plotting in second axes - the right one
axs2.barh(total_immigrants.index, total_immigrants) #Notice the use of 'barh' for creating horizontal bar plot
axs2.set_title("Bar plot on immigrants")

#Adding a Title for the Overall Figure
fig.suptitle('Subplotting Example', fontsize=15)

# Adjust spacing between subplots
fig.tight_layout()


# Show the figure
plt.show()
```

```python
# Set up the 2x2 subplot grid
fig, axs = plt.subplots(2, 2, figsize=(16, 12))
plt.subplots_adjust(wspace=0.3, hspace=0.4)  # Adjust spacing

# 1. Top Left: Immigration Trend for China and India (2000-2013)
years = [str(year) for year in range(2000, 2014)]
countries = ['China', 'India']
df_filtered = df_can[df_can['Country'].isin(countries)][['Country'] + years]
df_melted = df_filtered.melt(id_vars='Country', var_name='Year', value_name='Immigrants')

for country in countries:
    data = df_melted[df_melted['Country'] == country]
    axs[0,0].plot(data['Year'], data['Immigrants'], 'o-', label=country)
axs[0,0].set_title('1. China & India Immigration (2000-2013)')
axs[0,0].set_ylabel('Number of Immigrants')
axs[0,0].legend()
axs[0,0].grid(axis='y', linestyle='--', alpha=0.7)

# 2. Top Right: 5 Countries with Least Immigration
df_can['Total'] = df_can[years].sum(axis=1)
bottom_5 = df_can.nsmallest(5, 'Total')
bars = axs[0,1].bar(bottom_5['Country'], bottom_5['Total'], color='lightcoral')
axs[0,1].set_title('2. 5 Least Contributing Countries')
axs[0,1].set_ylabel('Total Immigrants (1980-2013)')
for bar in bars:
    height = bar.get_height()
    axs[0,1].text(bar.get_x() + bar.get_width()/2., height,
                 f'{int(height)}', ha='center', va='bottom')
axs[0,1].grid(axis='y', linestyle='--', alpha=0.4)

# 3. Bottom Left: Haiti Immigration (Line Plot)
haiti = df_can[df_can['Country'] == 'Haiti'][years].T
haiti.index = haiti.index.astype(int)
axs[1,0].plot(haiti.index, haiti.values, 'g-^')
axs[1,0].set_title('3. Haiti Immigration (1980-2013)')
axs[1,0].set_xlabel('Year')
axs[1,0].set_ylabel('Number of Immigrants')
axs[1,0].grid(True, linestyle=':', alpha=0.7)

# 4. Bottom Right: Continent Distribution (Pie Chart)
# Assuming df_con contains continent totals with 'Total' and index has continent names
axs[1,1].pie(df_con['Total'],
             colors=['gold','blue','lightgreen','coral','cyan','red'],
             autopct='%1.1f%%',
             pctdistance=0.8)
axs[1,1].set_title('4. Continent-Wise Distribution')
axs[1,1].legend(df_con.index, bbox_to_anchor=(1.3, 0.5))

plt.tight_layout()
plt.show()
```
